```python
def get_supplier_code(part_code):
    for i in range(len(part_code)):
        if part_code[i] == ":":
            return part_code[:i]
```

```python
print(get_supplier_code("ACME:123-L"))
```

```python
def get_product_num(part_code):
    colon_index = part_code.find(":")
    dash_index = part_code.find("-")
    return part_code[colon_index+1:dash_index]
```

```python
print(get_product_num("ACME:123-L"))
```

```python
def get_size(part_code):
    dash_index = part_code.find("-")
    return part_code[dash_index+1:]
```

```python
print(get_size("ACME:123-L"))
```

```python
def get_supplier_code(part_code):
    colon_index = part_code.find(":")
    return part_code[:colon_index]

                   
```

```python
def get_product_num(part_code):
    colon_index = part_code.find(":")
    dash_index = part_code.find("-")
    return part_code[colon_index+1:dash_index]
```

```python
def get_size(part_code):
    dash_index = part_code.find("-")
    return part_code[dash_index+1:]
```

```python
part1 = "ACME:123-L"
part2 = "DI:12345-M"
part3 = "ACE:1-12"
```

```python
print(f"For {part1}, the supplier is {get_supplier_code(part1)}. Part number {get_product_num(part1)} is available in size {get_size(part1)}.")
```

```python
part_list = ["ACME:123-L", "DI:12345-M", "ACE:1-12"]
```
