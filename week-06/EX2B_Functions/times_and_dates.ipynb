## Lab 2: Exercise 2.B
This notebook works with datetime functions.


```python
import datetime
```

#Today

```python
today = datetime.datetime.now()
print(today)
```

```python
print(today.strftime("Today is %A, %B %d, %Y"))
```

#My birthday

```python
birthday = datetime.date(2004, 5, 22)
```

```python
print("My birthday is", birthday.strftime("%m/%d/%Y")) 
```

# Next 90 days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(ninety_d.strftime("90 days from today is %B %d, %Y"))
```

#Time for Dinner 

```python
dinner_time = datetime.time(19,0)
```

```python
print(dinner_time.strftime("Lets meet for dinner at %I:%M %p"))
```
