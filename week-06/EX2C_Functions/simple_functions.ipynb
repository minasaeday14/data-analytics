This lab works with defining custom functions.


```python
def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}. \nMy favorite musician or band is {music}."
```

```python
my_name = "Mina"
my_movie = "Spider Man"
my_music = "Amanuel"
print(favorite_things(my_name, my_movie,my_music))
```

```python
def why_im_here(purpose):
    return purpose
```

```python
my_purpose = "I joined the program to build my data skills"
print(why_im_here(my_purpose))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
place_1 = "Chicago"
why_1 = "I enjoy the food and the downtown view"

print(favorite_place(place_1,why_1))
```

```python
place_2 = "Ethiopia"
why_2 = "My family lives there, and its a calming country"

print(favorite_place(place_2,why_2))
```
