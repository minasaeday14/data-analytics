## Tax Functions Lab
This lab demonstrateshow to create custom functions that calculate Social Security tax, Medicare tax, and federal tax withholding based on gross pay and withholding code.

```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

```python
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        tax_rate = 0.23
    if withholding_code == 1:
        tax_rate == 0.21
    if witholding_code == 2:
        tax_rate == 0.195
    elif withholding_code == 3:
        tax_rate = 0.185
    else:
        tax_rate = 0.18 - ((withholding_code - 4) * 0.005)

    return gross_pay * tax_rate
```

```python
gross_pay_1 = 750
withholding_code_1 = 0

print("Person 1")
print("Social Security Tax:", get_soc_sec_tax(gross_pay_1))
print("Medicare Tax:", get_medicare_tax(gross_pay_1))
print("Federal Tax:", get_federal_tax(gross_pay_1, withholding_code_1))
```
